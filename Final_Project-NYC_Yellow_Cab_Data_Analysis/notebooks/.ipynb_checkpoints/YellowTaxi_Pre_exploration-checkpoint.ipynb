```python
ride_estimate_url = "https://api.lyft.com/v1/cost"
pickup_longitude = "-73.981"
pickup_latitude = "40.730"

dropoff_longitude = "-73.988586"
dropoff_latitude = "40.758839"

payload = {'start_lat':pickup_latitude, 'start_lng': pickup_longitude,
           'end_lat': dropoff_latitude, 'end_lng': dropoff_longitude}

hdrs = {'Authorization': 'bearer '+access_token}

ride_estimate = requests.get(ride_estimate_url, params=payload, headers = hdrs)
ride_estimate
```

```python
#print(ride_estimate.json())
print(json.dumps(ride_estimate.json(), indent=2))
```

```python
data = ride_estimate.json()
```
