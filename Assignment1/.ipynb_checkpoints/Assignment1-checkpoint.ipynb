## Question1

```python
# Imports used in the assignemnt 1.
# to read a file and extract the path
import glob
import os
# To select random values from dictionary
import random
```

```python
L = []
# Use range function to create a list of numbers from 1 to 500
for x in range(1,500):
    if x%3 == 0 or x%7 == 0 :
        L.append(x)

print('Multiples of 3 and 7 are:')
print(L)
```

```python
# Mean = sum of elements in list divide by number of items on list
listLength = len(L)
mean = sum(L)/ listLength
print('Mean is {0}'.format(mean))

# Quartile calculations
lowerQuartile = (listLength)//4
print('25 Quartile is: {0}'.format(L[lowerQuartile]))

value = listLength//2
secondQuartile = (L[value-1] + L[value])/2
print('50 Quartile is nothing but the median of the list is : {0}'.format(secondQuartile))

upperQuartile = (3*(listLength)//4)
print('75 Quartile is: {0}'.format(L[upperQuartile]))
```

## Question 2

```python
# Read the file
input = open('ques2.txt')
aString = input.read()
# split using space
words = aString.split(' ')
print('Number of words is: {0}'.format(len(words)))
```

```python
# Create a list containing length per word
lengthPerWord = []
for x in words:
    lengthPerWord.append(len(x))

# List should be sorted for median calculation
lengthPerWord.sort()
lengthOfList = len(lengthPerWord)
# Average is sum of length word divide by total number of words.
# Length is excluding the space
print('Average length per word: {0}'.format(sum(lengthPerWord)/lengthOfList))
medianNumber = (lengthOfList+1)//2
print('Median for word length: {0}'.format(lengthPerWord[medianNumber]))
```

```python
#Create a dictionary
occuranceTable = {}
for i in lengthPerWord:
    occuranceTable[i] = occuranceTable.get(i, 0) + 1
# Print the key having max value which is mode
maxValKey = max(occuranceTable, key=occuranceTable.get)
print('Mode for word length is: {0}'.format(maxValKey))
```

## Question 3

```python
# Get the list of file names from the directory
fileNames = glob.glob('jsons/*.json')
i = 0
for fileName in fileNames:
    for line in open(fileName):
        # Increment the count if string is present in the line
        if 'user not found'.lower() in line.lower():
            i = i+1
            
print('Total count for word is: {0}'.format(i))
```

```python
fileTable = {}
for fileName in fileNames:
    # Extract userId
    userId = os.path.splitext(os.path.basename(fileName))[0]
    # Reading whole file in a string here
    if 'user not found'.lower() in open(fileName).read().lower():
        fileTable[userId] = 0
    else:
        fileTable[userId] = 1
        
print('Total files traversed: {0}'.format(len(fileTable)))

#Get 5 sample keys using random
sampleKeys = random.sample(list(fileTable), 5)

for key in sampleKeys:
    print('{0} , {1}'.format(key, fileTable[key]))
```

## Question 4

```python
import json
segmentTable = {}
for fileName in fileNames:
    # Open each file
    with open(fileName) as data_file:
        # Extract userId
        userId = os.path.splitext(os.path.basename(fileName))[0]
        # Make an entry with 0 values in dictionary this will be a nested dictionary
        segmentTable[userId] = {'first':0,'third':0}
        # load json file into data object
        data = json.load(data_file)
        # segmentUid contains if dataNode retargetingSegments contains in the json file
        if 'retargetingSegments' in data:
            # For each segmentUid make an entry in segmentTable
            for dataNode in data['retargetingSegments']:
                segmentUid = dataNode['segmentUid']
                if segmentUid.startswith('0'):
                    segmentTable[userId]['first'] += 1
                elif segmentUid.startswith('1'):
                    segmentTable[userId]['third'] += 1

# Print the directory
for key in segmentTable.keys():
    print('UserId: {0}\t\t first: {1}\t third: {2}'.format(key,segmentTable[key]['first'],segmentTable[key]['third']))
```
